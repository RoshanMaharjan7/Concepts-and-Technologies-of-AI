##Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
# Calculating the IQR for score
score_Q1 = south_asian_df['score'].quantile(0.25)
score_Q3 = south_asian_df['score'].quantile(0.75)
score_IQR = score_Q3 - score_Q1

# Calculating the IQR for GDP
GDP_Q1 = south_asian_df['Log GDP per capita'].quantile(0.25)
GDP_Q3 = south_asian_df['Log GDP per capita'].quantile(0.75)
GDP_IQR = GDP_Q3 - GDP_Q1

# Calculating lower and higher bound for score
lower_bound_score = score_Q1 - 1.5 * score_IQR
higher_bound_score = score_Q3 + 1.5 * score_IQR

# Calculating lower and higher bound for GDP
lower_bound_GDP = GDP_Q1 - 1.5 * GDP_IQR
higher_bound_GDP = GDP_Q3 + 1.5 * GDP_IQR

# Finding the outliers from the dataset
outliers = south_asian_df[(south_asian_df['score'] < lower_bound_score ) | (south_asian_df['score'] > higher_bound_score) |
                         (south_asian_df['Log GDP per capita'] < lower_bound_GDP) | (south_asian_df['Log GDP per capita'] > higher_bound_GDP)]

# Display
print("Outliers in the South Asian dataset:")
outliers
```

2. Define outliers using the 1.5 Ã— IQR rule.

```python
# SOUTH ASIA OUTLIERS
south_asia_score_Q1 = south_asian_df['score'].quantile(0.25)
south_asia_score_Q3 = south_asian_df['score'].quantile(0.75)
south_asia_score_IQR = south_asia_score_Q3 - south_asia_score_Q1

south_asia_GDP_Q1 = south_asian_df['Log GDP per capita'].quantile(0.25)
south_asia_GDP_Q3 = south_asian_df['Log GDP per capita'].quantile(0.75)
south_asia_GDP_IQR = south_asia_GDP_Q3 - south_asia_GDP_Q1

south_asia_lower_bound_score = south_asia_score_Q1 - 1.5 * south_asia_score_IQR
south_asia_higher_bound_score = south_asia_score_Q3 + 1.5 * south_asia_score_IQR

south_asia_lower_bound_GDP = south_asia_GDP_Q1 - 1.5 * south_asia_GDP_IQR
south_asia_higher_bound_GDP = south_asia_GDP_Q3 + 1.5 * south_asia_GDP_IQR

south_asia_outliers = south_asian_df[(south_asian_df['score'] < south_asia_lower_bound_score) | (south_asian_df['score'] > south_asia_higher_bound_score) | (south_asian_df['Log GDP per capita'] < south_asia_lower_bound_GDP) | (south_asian_df['Log GDP per capita'] > south_asia_higher_bound_GDP)]

print("South Asia Outliers based on Score and GDP per Capita:")
print(south_asia_outliers)



# MIDDLE EAST OUTLIERS
middle_east_score_Q1 = middle_east_df['score'].quantile(0.25)
middle_east_score_Q3 = middle_east_df['score'].quantile(0.75)
middle_east_score_IQR = middle_east_score_Q3 - middle_east_score_Q1

middle_east_GDP_Q1 = middle_east_df['Log GDP per capita'].quantile(0.25)
middle_east_GDP_Q3 = middle_east_df['Log GDP per capita'].quantile(0.75)
middle_east_GDP_IQR = middle_east_GDP_Q3 - middle_east_GDP_Q1

middle_east_lower_bound_score = middle_east_score_Q1 - 1.5 * middle_east_score_IQR
middle_east_higher_bound_score = middle_east_score_Q3 + 1.5 * middle_east_score_IQR

middle_east_lower_bound_GDP = middle_east_GDP_Q1 - 1.5 * middle_east_GDP_IQR
middle_east_higher_bound_GDP = middle_east_GDP_Q3 + 1.5 * middle_east_GDP_IQR

middle_east_outliers = middle_east_df[(middle_east_df['score'] < middle_east_lower_bound_score) | (middle_east_df['score'] > middle_east_higher_bound_score) | (middle_east_df['Log GDP per capita'] < middle_east_lower_bound_GDP) | (middle_east_df['Log GDP per capita'] > middle_east_higher_bound_GDP)]

print("\nMiddle East Outliers based on Score and GDP per Capita:")
print(middle_east_outliers)

```
